```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT805/Assignment 2/Results/average_rating_by_time.csv', '\t')
d.columns = ['hour', 'rating']
d.loc[d.shape[0]] = [8, 1843]
d.head()
```

```python
d = d.sort_values(by=['hour'],ascending=True)
d
```

```python
plt.rcParams['figure.dpi'] =150
plt.plot(d['hour'], d['rating'], c = 'blue')
plt.title('Average Rating By Hour-of-Day')
plt.xlabel('Hour-of-Day')
plt.ylabel('Average Rating')
```

```python
d = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT805/Assignment 2/Results/average_rating_by_event.csv', '\t')
d.columns = ['event', 'rating']
d.loc[d.shape[0]] = ['classical_tournament',1849]
d = d.sort_values(by=['rating'],ascending=False)
d
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
event = d['event']
rating = d['rating']
ax.bar(event,rating, color = 'black')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating By Game Type')
plt.show()
```
